```python
import pandas as pd
```

```python
community_df=pd.read_csv('PALS1.txt', header=None, delimiter= "\n")
```

```python
edges_df=pd.read_csv('COLLMSGLINK1.csv', header=None, delimiter= ",")
```

```python
#edges_df
```

```python
# Create a time dictionary with time as key and src nodes as values. Each src node is a dictionary of target nodes
time_dict = {}
for i in range(1, len(edges_df)):
    src_node = edges_df[0][i]
    tgt_node = edges_df[1][i]
    time = edges_df[2][i]
    if time not in time_dict:
        time_dict[time] = {}
    if src_node not in time_dict[time]:
        time_dict[time][src_node] = set()
    time_dict[time][src_node].add(tgt_node) 
#print(time_dict)  # Time: src: [tgt nodes]
#print(time_dict['0']['1'])
```

```python
occ_of_time = {}
count = 0
lines_of_time = []
for i in range(len(community_df)):
    if community_df[0][i].startswith('Time'):
        occ_of_time[i] = count
        lines_of_time.append(i)
        count += 1
#print(occ_of_time)
```

```python
# Create a list of lists to store time, community and src nodes
rows = []
#print(lines_of_time)
current_time = -1
for i in range(len(community_df)):
    if i in lines_of_time:
        current_time += 1
        community_count = 0
    if community_df[0][i].startswith('Community'):
        i += 1
        no_of_nodes = community_df[0][i]
        i += 1
        src_nodes = community_df[0][i]
        row = [current_time, community_count, int(no_of_nodes), src_nodes]
        community_count += 1
        rows.append(row)
#print(rows)  # Time, Community, No of Nodes, Nodes  
```

```python
# Function to convert string elements to int elements
relative_edges_data = []
def string_to_int_list(test_list):
    b_test_list = []
    a_test_list = test_list[1:len(test_list) - 1]
    a_test_list = list(a_test_list.split(', '))
    for x in a_test_list:
        b_test_list.append(int(x.strip()))
    return b_test_list
for i in rows:
    time = str(i[0])
    community = i[1]
    no_of_nodes = i[2]
    src_nodes = i[3]
    src_nodes = string_to_int_list(src_nodes)
    edges = 0
    for src_node in src_nodes:
        src_node = str(src_node)
        if time not in time_dict:
            continue
        if src_node not in time_dict[time]:
            continue
        edges = edges + len(time_dict[time][src_node])
    #print('Edges at Time ', time, ' in community ', community, ' is ', edges)
    relative_edge_data = [time, community, no_of_nodes, edges]
    relative_edges_data.append(relative_edge_data)
    
#print(relative_edges_data)
```

```python
total_edges = []
total = int(relative_edges_data[0][3])
for i in range(1, len(relative_edges_data)):
    total += int(relative_edges_data[i][3])
    if int(relative_edges_data[i][0]) == int(relative_edges_data[i-1][0]) + 1 or i == len(relative_edges_data) - 1:
        total_edges.append(total - relative_edges_data[i][3])
        if i == len(relative_edges_data) - 1:
            total_edges.pop()
            total_edges.append(total)
        total = relative_edges_data[i][3]
#print(total_edges)
```

```python
#Calculation of relative edges and density
relative_edges = []
for i in range(len(relative_edges_data)):
    time = int(relative_edges_data[i][0])
    community = relative_edges_data[i][1]
    no_of_nodes = relative_edges_data[i][2]
    community_edges = relative_edges_data[i][3]
    relative_edge = community_edges/total_edges[time]
    relative_edge = round(relative_edge, 3)
    density = (community_edges * 2) / (no_of_nodes * (no_of_nodes - 1))
    density = round(density, 3)
    #print(relative_edge)
    relative_edges.append([time, community, community_edges, no_of_nodes, relative_edge, density])
#     print(round(relative_edges, 3))

#print(relative_edges)
```

```python
#Calculation of relative_edges
df = pd.DataFrame(relative_edges, columns = ['Time','Community','Community_Edges','No_Of_Nodes','Relative_Edges','Density'])
df
```

```python
# Saving relative edges
df.to_csv('relative_edge_result1.csv', index = False)  #columns=['S_No','Time','Community','Community_Edges','No_Of_Nodes','Relative_Edges','Density']
```
